```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy import signal
from scipy.io import wavfile
```

```python
sns.set()
```

```python
sample_rate, samples = wavfile.read('akcie_komercni_banky.wav')
frequencies, times, spectrogram = signal.spectrogram(samples, sample_rate)
```

```python
plt.pcolormesh(times, frequencies, spectrogram)
plt.imshow(np.log(spectrogram))
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```

```python
print(sample_rate)
plt.specgram(samples, Fs=sample_rate, cmap='RdBu_r')
plt.ylabel('Frekvence [Hz]')
plt.xlabel('as [s]')
plt.savefig('spectrogram_kout.png', 
            dpi=300, # Dots per inch
            # frameon='false',
            aspect='normal',
            bbox_inches='tight',
            pad_inches=0) 
plt.show()
```

```python
normal_sample_rate, normal_samples = wavfile.read('akcie_komercni_banky_normal.wav')

print(normal_sample_rate)

plt.specgram(normal_samples, Fs=normal_sample_rate, cmap='RdBu_r')
plt.ylabel('Frekvence [Hz]')
plt.xlabel('as [s]')
plt.savefig('spectrogram_normal.png', 
            dpi=300, # Dots per inch
            # frameon='false',
            aspect='normal',
            bbox_inches='tight',
            pad_inches=0) 
plt.show()
```

```python
ax1 = plt.subplot(1, 2, 1)
plt.specgram(normal_samples, Fs=normal_sample_rate, cmap='RdBu_r')
plt.ylabel('Frekvence [Hz]')
plt.xlabel('as [s]')

ax2 = plt.subplot(1, 2, 2, sharey=ax1)
plt.specgram(samples, Fs=sample_rate, cmap='RdBu_r')
plt.xlabel('as [s]')
plt.setp(ax2.get_yticklabels(), visible=False)

plt.savefig('../img/spectrogram_subplot.png', 
            dpi=300, # Dots per inch
            # frameon='false',
            aspect='normal',
            bbox_inches='tight',
            pad_inches=0)

plt.show()
```
